```python
from pathlib import Path

p = Path("EDGE.py")
text = p.read_text()

old = """checkpoint = torch.load(
                checkpoint_path, map_location=self.accelerator.device
            )"""

new = """checkpoint = torch.load(
                checkpoint_path, map_location=self.accelerator.device, weights_only=False
            )"""

if old not in text:
    print("Original pattern not found. Showing nearby torch.load lines:")
    for i, line in enumerate(text.splitlines(), start=1):
        if "torch.load" in line:
            print(i, line)
else:
    text = text.replace(old, new)
    p.write_text(text)
    print("patched EDGE.py")
```

```python
import re
import csv
import shutil
import subprocess
from pathlib import Path

motion_dir = Path("motions_for_pfc")
tmp_dir = Path("pfc_single_tmp")

# 清理临时文件夹
if tmp_dir.exists():
    shutil.rmtree(tmp_dir)
tmp_dir.mkdir(parents=True, exist_ok=True)

pkl_files = sorted(motion_dir.glob("*.pkl"))

print("Found pkl files:")
for f in pkl_files:
    print(" -", f)

rows = []

for f in pkl_files:
    # 每次只放一个 pkl 文件进去
    for old in tmp_dir.glob("*"):
        old.unlink()

    shutil.copy(f, tmp_dir / f.name)

    output = subprocess.check_output(
        ["python", "eval/eval_pfc.py", "--motion_path", str(tmp_dir)],
        text=True
    )

    print("\n=====", f.name, "=====")
    print(output.strip())

    match = re.search(r"mean PFC of ([0-9.]+)", output)
    pfc = float(match.group(1)) if match else None

    name = f.stem.replace("test_", "")

    if name.startswith("beat"):
        music_type = "strong beat"
    elif name.startswith("slow"):
        music_type = "slow music"
    elif name.startswith("speech"):
        music_type = "speech / weak beat"
    else:
        music_type = "other"

    rows.append({
        "motion_file": f.name,
        "music_file": name + ".wav",
        "music_type": music_type,
        "pfc": pfc
    })

# 保存 CSV
with open("pfc_by_file.csv", "w", newline="") as out:
    writer = csv.DictWriter(
        out,
        fieldnames=["motion_file", "music_file", "music_type", "pfc"]
    )
    writer.writeheader()
    writer.writerows(rows)

print("\nSaved to pfc_by_file.csv")
```
